## SPLINES

```python
import numpy as np
import matplotlib.pyplot as plt

x_s = np.array([1, 2, 3])
y_s = np.array([1, 5, 11])
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_s,y_s)
plt.xlabel("x_s")
plt.ylabel("y_s")
plt.title(" Dataset ")
plt.show()
```

```python
def S0(x):
    b0 = 3.5  
    d0 = 0.5  
    return 1 + b0 * (x - 1) + d0 * (x - 1)**3

def S1(x):
    b1 = 5    
    c1 = 1.5 
    d1 = -0.5 
    return 5 + b1 * (x - 2) + c1 * (x - 2)**2 + d1 * (x - 2)**3
```

```python
x_values_0 = np.linspace(1, 2, 100)  
x_values_1 = np.linspace(2, 3, 100)

y_values_0 = S0(x_values_0)
y_values_1 = S1(x_values_1)
```

```python
plt.plot(x_values_0, y_values_0, label="Spline S0(x)")
plt.plot(x_values_1, y_values_1, label="Spline S1(x)")
plt.scatter(x_s, y_s, color="red", label="Puntos dados")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Spline CÃºbico")
plt.legend()
plt.grid()
plt.show()
```
