```python
#graficar 
x_s = [1, 2, 3]
y_s = [1, 2, 2]
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_s,y_s)
plt.xlabel("x_s")
plt.ylabel("y_s")
plt.title("Gráfica")
plt.show()
```

```python
def f(x:float) -> float:
    return -(x**2)/2 + 5*x / 2 -1
```

```python
import numpy as np

x_values = np.linspace(min(x_s),4,100)
y_values = [f(x) for x in x_values]

plt.plot(x_values,y_values,label = "f(x)")
plt.scatter(x_s,y_s,color = "red")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x)")
plt.legend()
```
