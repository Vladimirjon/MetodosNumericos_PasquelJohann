## LAGRANGE

```python
import numpy as np
import matplotlib.pyplot as plt


x_s1 = np.array([-2, 1, 3, 7])
y_s1 = np.array([5, 7, 11, 34])
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_s1,y_s1)
plt.xlabel("x_s")
plt.ylabel("y_s")
plt.title("- Puntos a usar -")
plt.show()
```

```python
def lagrange_polynomial(x: float, x_points: np.ndarray, y_points: np.ndarray) -> float:
    n = len(x_points)
    result = 0
    for k in range(n):
        L_k = 1
        for i in range(n):
            if i != k:
                L_k *= (x - x_points[i]) / (x_points[k] - x_points[i])
        result += y_points[k] * L_k
    return result
```

```python
x_values = np.linspace(min(x_s1), max(x_s1), 500)
y_values = [lagrange_polynomial(x, x_s1, y_s1) for x in x_values]
```

```python
# Grafica de puntos x_1 & y_1
plt.plot(x_values, y_values, label="Polinomio de Lagrange")
plt.scatter(x_s1, y_s1, color="red", label="Puntos dados")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Polinomio de Lagrange que pasa por los puntos dados")
plt.legend()
plt.grid()
plt.show()
```
