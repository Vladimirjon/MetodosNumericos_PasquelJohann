## MÃ­nimos Cuadrados - Pol. 2do Grado

```python
import matplotlib.pyplot as plt
import numpy as np

def equation(x):
    return 0.25 * x**2 -x - 1

x_values = np.linspace(-8, 8, 200)
y_values = equation(x_values)

x_s1 = np.array([-2, 0, 2, 6])
y_s1 = np.array([2, -1, -2, 2])

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='y = 0.25 x^2 - x - 1')

plt.scatter(x_s1, y_s1, color='red', label='Puntos adicionales')

plt.xlabel('x')
plt.ylabel('y')
plt.title('MIN CUADRADOS')
plt.legend()
plt.grid(True)

plt.show()
```
