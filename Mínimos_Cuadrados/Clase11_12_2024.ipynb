## MÃ­nimos Cuadrados

```python
import numpy as np
import matplotlib.pyplot as plt


x_s1 = np.array([-3,-1,2])
y_s1 = np.array([4.25,4.75,5.5])
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_s1,y_s1)
plt.xlabel("x_s")
plt.ylabel("y_s")
plt.title("- Puntos a usar -")
plt.show()
```

```python
def equation(x):
    return 0.25*x + 5
```

```python
x_values = np.linspace(-15, 15, 400)  
y_values = equation(x_values)

plt.plot(x_values, y_values, label='y = 0.25x + 5')
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = 0.25x + 5')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def equation(x):
    return 0.25 * x + 5

x_values = np.linspace(-3, 3, 400)  
y_values = equation(x_values)

x_s1 = np.array([-3, -1, 2])
y_s1 = np.array([4.25, 4.75, 5.5])

plt.plot(x_values, y_values, label='y = 0.25x + 5')

plt.scatter(x_s1, y_s1, color='red', label='Puntos adicionales')

plt.xlabel('x')
plt.ylabel('y')
plt.title('mÃ­nimos cuadrados')
plt.legend()
plt.grid(True)

plt.show()
```
