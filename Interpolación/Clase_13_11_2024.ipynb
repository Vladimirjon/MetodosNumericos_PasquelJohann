F(x) = cos(x)

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Equation F(x) = cos(x)
def equation(x):
    return np.cos(x)

def taylor_series_cos(x, terms):
    taylor_sum = np.zeros_like(x)
    for n in range(terms):
        term = ((-1)**n) * (x**(2 * n)) / math.factorial(2 * n)
        taylor_sum += term
    return taylor_sum

def plot_cos_with_taylor(x, max_terms):
    y = equation(x)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='F(x) = cos(x)', color='blue')
    for terms in range(1, max_terms + 1):
        y_taylor = taylor_series_cos(x, terms)
        plt.plot(x, y_taylor, '--', label=f'Término {terms}: Taylor con {terms} términos')
    
    # x_0 = 0
    plt.plot(0, equation(0), 'ro', label='x_0 = 0')

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Gráfica de F(x) = cos(x) y aproximaciones de Taylor hasta {max_terms} términos')
    plt.ylim([-3, 3])
    plt.grid(True)
    plt.legend()
    plt.show()


x = np.linspace(-6, 6, 100)
plot_cos_with_taylor(x, 3)
```

F(x) = ln(x)

```python
import numpy as np
import matplotlib.pyplot as plt

# Equation F(x) = ln(x)
def equation(x):
    return np.log(x)

def taylor_series_ln(x, terms):
    taylor_sum = np.zeros_like(x)
    for n in range(1, terms + 1):
        term = ((-1)**(n + 1)) * ((x - 1)**n) / n
        taylor_sum += term
    return taylor_sum

def plot_ln_with_taylor(x, max_terms):
    y = equation(x)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='F(x) = ln(x)', color='blue')
    
    for terms in range(1, max_terms + 1):
        y_taylor = taylor_series_ln(x, terms)
        plt.plot(x, y_taylor, '--', label=f'Término {terms}: Taylor con {terms} términos')
    
    # x_0 = 1
    plt.plot(1, equation(1), 'ro', label='x_0 = 1')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Gráfica de F(x) = ln(x) y aproximaciones de Taylor hasta {max_terms} términos')
    plt.ylim([-2, 2])
    plt.xlim([0.1, 3])
    plt.grid(True)
    plt.legend()
    plt.show()

x = np.linspace(0.1, 3, 100)
plot_ln_with_taylor(x, 3)
```

F(x) = $\frac{1}{1 - x}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Equation F(x) = 1 / (1 - x)
def equation(x):
    return 1 / (1 - x)

def taylor_series_1_over_1_minus_x(x, terms):
    taylor_sum = np.zeros_like(x)
    for n in range(terms):
        term = x**n
        taylor_sum += term
    return taylor_sum

def plot_1_over_1_minus_x_with_taylor(x, max_terms):
    y = equation(x)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='F(x) = 1/(1 - x)', color='blue')
    
    for terms in range(1, max_terms + 1):
        y_taylor = taylor_series_1_over_1_minus_x(x, terms)
        plt.plot(x, y_taylor, '--', label=f'Término {terms}: Taylor con {terms} términos')
    
    # x_0 = 0
    plt.plot(0, equation(0), 'ro', label='x_0 = 0')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Gráfica de F(x) = 1/(1 - x) y aproximaciones de Taylor hasta {max_terms} términos')
    plt.ylim([-5, 5])
    plt.xlim([-0.9, 0.9])
    plt.grid(True)
    plt.legend()
    plt.show()

x = np.linspace(-5, 5, 100)
plot_1_over_1_minus_x_with_taylor(x, 5)
```
