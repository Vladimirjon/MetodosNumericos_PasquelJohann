```python
import numpy as np
import matplotlib.pyplot as plt

#Ecuation

def equation(x):
    return (x**3 - 7*x**2 + 4*x + 12)

x = np.linspace(-15,15,100)
y = equation(x)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Ecuacion')
ax = plt.gca()
ax.set_ylim([-15,15])
ax.set_xlim([-15,15])
plt.plot(x,y)
plt.grid(True)
plt.show()
```
