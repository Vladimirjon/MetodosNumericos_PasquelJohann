```python
def newton_with_tracking(f, df, x0, tol=1e-6, max_iter=100):
    x = x0
    newton_approximations = [x]  
    for i in range(max_iter):
        x_new = x - f(x) / df(x) 
        newton_approximations.append(x_new)
        if abs(x_new - x) < tol:  
            break
        x = x_new
    return x, newton_approximations


x0 = -1
x_r, newton_approximations = newton_with_tracking(f, df, x0)


print(f"La raz encontrada por el mtodo de Newton es: x = {x_r}")
print("Aproximaciones de Newton:", newton_approximations)
```
