#### 1. $\frac{1}{25 \cdot x^2 + 1}$, $x_0 = 0$

#### Serie de Taylor

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Equation F(x) = 1 / (25 * x^2 + 1)
def equation(x):
    return 1 / (25 * x**2 +1)
```

```python
def taylor_series_f(x, terms):
    taylor_sum = np.zeros_like(x)
    for n in range(terms):
        coefficient = (-1)**n * math.factorial(2 * n) / (math.factorial(n) * (25**n))
        term = coefficient * x**(2 * n)
        taylor_sum += term
    return taylor_sum
```

```python
def plot_f_with_taylor(x, max_terms):
    y = equation(x)
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='F(x) = 1 / (25x¬≤ + 1)', color='blue')
    for terms in range(1, max_terms + 1):
        y_taylor = taylor_series_f(x, terms)
        plt.plot(x, y_taylor, '--', label=f'T√©rmino {terms}: Taylor con {terms} t√©rminos')
    
    # x_0 = 0
    plt.plot(0, equation(0), 'ro', label='x_0 = 0')

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Gr√°fica de F(x) = 1 / (25x¬≤ + 1) y aproximaciones de Taylor hasta {max_terms} t√©rminos')
    plt.ylim([-1, 2])
    plt.grid(True)
    plt.legend()
    plt.show()

# Intervalo de x y con 1 t√©rmino
x = np.linspace(-2, 2, 100)
plot_f_with_taylor(x, 1)
```

Para esta funci√≥n $F(x) = \frac{1}{25x^2 + 1}$, la Serie de Taylor **no proporciona una aproximaci√≥n pr√°ctica para valores lejanos a $x = 0$**. 

La mejor aproximaci√≥n usando Taylor ser√≠a con **pocos t√©rminos y en un rango muy cercano a $x = 0$** 

La Serie de Taylor **no es ideal** para esta funci√≥n. Esta funci√≥n es una **racional**, con una dependencia cuadr√°tica en el denominador $25x^2 + 1$. Cuando se a√±aden m√°s t√©rminos a la Serie de Taylor, los polinomios tienden a oscilar y divergir lejos de $x_0 = 0$.

#### Polinomio de Lagrange

En este caso, para construir el polinomio de Lagrange de $F(x) = \frac{1}{25x^2 + 1}$, usaremos diferentes cantidades de puntos para las distintas aproximaciones.


**Primera aproximaci√≥n** 

Puntos: $(0, 1)$ , $(-0.4, 0.2)$ 

```python
x_1aproximacion = [0,-0.4]
y_1aproximacion = [1, 0.2]
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_1aproximacion,y_1aproximacion)
plt.xlabel("x_1aproximacion")
plt.ylabel("y_1aproximacion")
plt.title("Gr√°fica")
plt.show()
```

```python
def lagrange_polynomial(x: float, x_points: np.ndarray, y_points: np.ndarray) -> float:
    n = len(x_points)
    result = 0
    for k in range(n):
        L_k = 1
        for i in range(n):
            if i != k:
                L_k *= (x - x_points[i]) / (x_points[k] - x_points[i])
        result += y_points[k] * L_k
    return result
```

```python
x_values_1aproximacion = np.linspace(min(x_1aproximacion), max(x_1aproximacion), 500)
y_values_1aproximacion = [lagrange_polynomial(x, x_1aproximacion, y_1aproximacion) for x in x_values_1aproximacion]
```

```python
plt.plot(x_values_1aproximacion, y_values_1aproximacion, label="Polinomio de Lagrange")
plt.scatter(x_1aproximacion, y_1aproximacion, color="red", label="Puntos dados")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Polinomio de Lagrange que pasa por los puntos dados")
plt.legend()
plt.grid()
plt.show()
```

#### 2. $arctanx$, $x_0 = 1$

#### Serie de Taylor

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# Equation F(x) = arctan(x)
def equation(x):
    return np.arctan(x)
```

```python
def taylor_series_arctan(x, terms):
    taylor_sum = np.zeros_like(x)  
    for n in range(terms):
        coefficient = (-1)**n / (2 * n + 1) 
        term = coefficient * x**(2 * n + 1)  
        taylor_sum += term
    return taylor_sum
```

```python
def plot_arctan_with_taylor(x, max_terms):
    y = np.arctan(x)  
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='F(x) = arctan(x)', color='blue')  
    
    for terms in range(1, max_terms + 1):
        y_taylor = taylor_series_arctan(x, terms)  
        plt.plot(x, y_taylor, '--', label=f'T√©rmino {terms}: Taylor con {terms} t√©rminos')
    
    # Punto x_0 = 1
    plt.plot(1, np.arctan(1), 'ro', label='x_0 = 1 (Valor exacto)')
    
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Gr√°fica de F(x) = arctan(x) y aproximaciones de Taylor hasta {max_terms} t√©rminos')
    plt.ylim([-1, 2])
    plt.grid(True)
    plt.legend()
    plt.show()


x = np.linspace(-2, 2, 100)
plot_arctan_with_taylor(x, 2)  # Ejemplo con 2 t√©rminos
```

**La serie de Taylor tiene un rango de convergencia limitado por su punto de expansi√≥n $x_0 = 1$**

El orden √≥ptimo depende del rango en el que se desee precisi√≥n y el nivel de error aceptable. Para aplicaciones pr√°cticas, evaluar el error puede ayudar a determinar cu√°ntos t√©rminos son suficientes.

A medida que incluimos t√©rminos de mayor orden, la acumulaci√≥n de errores num√©ricos o la complejidad de los coeficientes puede impactar ligeramente la precisi√≥n de la gr√°fica.

**La funci√≥n $arctan(x)$ tiene un crecimiento lento y la serie de Taylor no capta este comportamiento de forma eficiente con pocos t√©rminos.**

#### Polinomio de Lagrange

En este caso, para construir el polinomio de Lagrange de $arctanx$, usaremos diferentes cantidades de puntos para las distintas aproximaciones.

```python
def lagrange_polynomial(x: float, x_points: np.ndarray, y_points: np.ndarray) -> float:
    n = len(x_points)
    result = 0
    for k in range(n):
        L_k = 1
        for i in range(n):
            if i != k:
                L_k *= (x - x_points[i]) / (x_points[k] - x_points[i])
        result += y_points[k] * L_k
    return result
```
